```python
LIMIT = 2 ** 24
```

```python
def is_self_describing(n):
    s = str(n)

    # Examine if number is valid
    if len(s) % 2 != 0:
        return False
    
    expected_counts = [0] * 10

    # Store expected counts
    for i in range (0, len(s), 2):
        count = int(s[i])
        digit = int(s[i+1])
        expected_counts[digit] = count
    
    # Count actual occurrences of each digit
    actual_counts = [0] * 10

    for digit in s:
        actual_counts[int(digit)] += 1

    return actual_counts == expected_counts
```

```python
for d in range(1, LIMIT + 1):
    if is_self_describing(d):
        print(d)
```
